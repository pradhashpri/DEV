## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D contour plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(X, Y, Z, levels=50, cmap='viridis')
```

```
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Contour Plot')
plt.show()
import matplotlib
```

```
import matplotlib.pyplot as plt
from 'mpl_toolkits.basemap' import Basemap
import numpy as np

# Create a map
m = Basemap(projection='lcc', lat_0=35, lon_0=-100, width=1E6, height=1.2E6, resolution='i')

# Draw map features
m.drawcoastlines()
m.drawcountries()
m.drawstates()
# Generate synthetic temperature data
lons = np.linspace(-110, -90, 100)
lats = np.linspace(30, 45, 100)
lons, lats = np.meshgrid(lons, lats)
temps = np.sin(lats / 5.0) * np.cos(lons / 5.0) * 20 + 10  # Synthetic temperature data

# Convert to map projection coordinates
x, y = m(lons, lats)

# Plot the temperature data
cs = m.contourf(x, y, temps, cmap='coolwarm')
m.colorbar(cs, label='Surface Temperature (°C)')

plt.title('Surface Temperature')
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate sample data
np.random.seed(0)
data = {
    'age': np.random.randint(18, 70, 200),
    'gender': np.random.choice(['Male', 'Female'], 200),
    'finishing_time': np.random.normal(240, 30, 200)  # finishing time in minutes
}
df = pd.DataFrame(data)
# Scatter Plot: Age vs. Finishing Time
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='finishing_time', hue='gender', palette='deep')
plt.title('Age vs. Marathon Finishing Time')
plt.xlabel('Age')
plt.ylabel('Finishing Time (minutes)')
plt.legend(title='Gender')
plt.show()
```

```
# Simulate household survey data
household_data = {
    'Household Size': [3, 2, 4, 5, 1, 6],
    'Annual Income': [50000, 60000, 55000, 52000, 49000, 80000],
    'Region': ['Urban', 'Rural', 'Urban', 'Suburban', 'Rural', 'Urban']
}

df_household = pd.DataFrame(household_data)
print(df_household)
```

```
df_household['Annual Income'] = df_household['Annual Income'].round(-3)  # Round to nearest thousand
print(df_household)
```

```
# Bar chart for categorical variable 'Region'
plt.figure(figsize=(8, 6))
sns.countplot(x='Region', data=df_household)
plt.title('Household Regions')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```
# Pie chart for categorical variable 'Region'
region_counts = df_household['Region'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Household Regions')
plt.show()
```

```
# Histogram for numeric variable 'Annual Income'
plt.figure(figsize=(8, 6))
sns.histplot(df_household['Annual Income'], bins=5, kde=True)
plt.title('Household Annual Income Distribution')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.show()
```

```
# Recode 'Annual Income' into categories
bins = [0, 50000, 60000, 100000]  # Define bin edges
labels = ['Low', 'Medium', 'High']  # Define labels
df_household['Income Category'] = pd.cut(df_household['Annual Income'], bins=bins, labels=labels)
print(df_household)
```

```
import pandas as pd

# Simulated dataset of working hours
data = {'Hours Worked': [40, 42, 38, 45, 50, 41, 39, 47, 44, 36]}
df = pd.DataFrame(data)


# Calculate level (mean) and spread (standard deviation)
mean = df['Hours Worked'].mean()
std_dev = df['Hours Worked'].std()

print(f"Mean Hours Worked: {mean}")
print(f"Standard Deviation of Hours Worked: {std_dev}")
```

```
# Simulated dataset of couples' working hours
couple_data = {'Husband': [40, 42, 38, 48, 50], 'Wife': [35, 38, 32, 40, 42]}
df_couples = pd.DataFrame(couple_data)


# Summaries of couples' working hours
mean_husband = df_couples['Husband'].mean()
mean_wife = df_couples['Wife'].mean()

print(f"Average Husband's Hours: {mean_husband}")
print(f"Average Wife's Hours: {mean_wife}")
```

```
# Calculate mean, median, mode
median = df['Hours Worked'].median()
mode = df['Hours Worked'].mode()[0]

print(f"Median Hours Worked: {median}")
print(f"Mode Hours Worked: {mode}")
```

```
# Calculate range and interquartile range (IQR)
range_val = df['Hours Worked'].max() - df['Hours Worked'].min()
iqr = df['Hours Worked'].quantile(0.75) - df['Hours Worked'].quantile(0.25)

print(f"Range: {range_val}")
print(f"Interquartile Range (IQR): {iqr}")
```

```
# Calculate quartiles
q1 = df['Hours Worked'].quantile(0.25)
q3 = df['Hours Worked'].quantile(0.75)

print(f"Q1 (25th Percentile): {q1}")
print(f"Q3 (75th Percentile): {q3}")
```

```
# Aggregating working hours data
total_hours = df['Hours Worked'].sum()
mean_hours = df['Hours Worked'].mean()

print(f"Total Hours Worked: {total_hours}")
print(f"Mean Hours Worked: {mean_hours}")
# Simulated dataset of hours worked by age
age_data = {'Age': [22, 30, 45, 55, 60], 'Hours Worked': [35, 40, 50, 45, 38]}
df_age = pd.DataFrame(age_data)

# Visualizing the relationship between age and hours worked
import matplotlib.pyplot as plt
plt.scatter(df_age['Age'], df_age['Hours Worked'])
plt.title('Hours Worked by Age')
plt.xlabel('Age')
plt.ylabel('Hours Worked')
plt.show()
```

```
# Simulated dataset of working hours across countries
europe_data = {'Country': ['UK', 'Germany', 'France', 'Spain', 'Italy'], 'Hours Worked': [40, 38, 35, 36, 39]}
df_europe = pd.DataFrame(europe_data)

# Bar chart comparing working hours across Europe
plt.figure(figsize=(8,6))
plt.bar(df_europe['Country'], df_europe['Hours Worked'])
plt.title('Average Weekly Working Hours Across Europe')
plt.xlabel('Country')
plt.ylabel('Hours Worked')
plt.show()
```
